```python
from Embed import *
import json
import networkx as nx
from networkx.readwrite import json_graph
import pandas as pd
import time


from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
#from sklearn.linear_model import 
from sklearn.metrics import accuracy_score

import os
import numpy as np
```

```python
f = "./data/git_stargazers/git_edges.json"
c = 0 
labels = pd.read_csv('./data/git_stargazers/git_target.csv')
#df = pd.DataFrame()
data = []
temp = []
info = []
LEN = []
with open(f) as f:
    for obj in f:
        gr = json.loads(obj)
        #gr = sorted(gr)
        #k = gr
        #print(k[0:5])
        for id,g in gr.items():#.values():
            lines = []
            print(id)
            #print(g)
            for edge in g:
                lines.append(f"{edge[0]}  {edge[1]}")
            c+=1
            print(c)
        #js_graph = json.load(gr)
        #G = json_graph.node_link_graph(gr)
            G = nx.parse_edgelist(lines)
            g = nodes_embedding(G)
            embedding , T,leng= graph_embedding(G,level=3,alpha=.9)
            #print(embedding)
            info.append([len(G.edges()),len(G.nodes())])
            data.append(embedding)
            temp.append(T)
            LEN.append(leng)
            #Stime.sleep(30)
            #print(leng)
            if(c>100) :
                break
        print('read objects done !')
        #time.sleep(30)
        
pd.DataFrame(data)            
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.75)
len(X_train)
```

```python
pred = clf.predict(X_test)
accuracy_score(pred,y_test)
```
